```python
bios['height_category'] = bios['height_cm'].apply(lambda x: "Short" if x < 160 else ("Average" if x < 180 else "Tall"))
```

```python
bios['height_category'].value_counts()
```

```python
def categorise_data(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 175 and row['weight_kg'] > 70:
        return "Balanced"
    else:
        return "Heavyweight"
```

```python
bios['Category'] = bios.apply(categorise_data, axis = 1)
```
